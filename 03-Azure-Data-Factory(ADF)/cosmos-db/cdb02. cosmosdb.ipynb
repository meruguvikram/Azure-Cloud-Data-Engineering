```python
# relace below url and primerkey

import os
from azure.cosmos import CosmosClient, exceptions, PartitionKey
#from azure.cosmos.aio import CosmosClient 
import json
import asyncio

URL = "https://rgrrcdb.documents.azure.com:443/"
KEY = "<Primerkey>"
client = CosmosClient(URL, credential=KEY)
```

```python
## Create a database
DATABASE_NAME='scott'
try:
    database = client.create_database(DATABASE_NAME)
except exceptions.CosmosResourceExistsError:
    database = client.get_database_client(DATABASE_NAME)
```

```python
# verify Database created or not
for i in client.list_databases():
    print(json.dumps(i,indent=4))
```

```python
## Create a container
CONTAINER_NAME = 'emp'
try:
    container = database.create_container(id=CONTAINER_NAME, partition_key=PartitionKey(path='/id', kind='Hash'))
except exceptions.CosmosResourceExistsError:
    container = database.get_container_client(CONTAINER_NAME)
except exceptions.CosmosHttpResponseError:
    raise
```

```python
# verify container created or not
for i in database.list_containers():
    print(json.dumps(i,indent=4))
```

```python
## Create records
for i in range(1, 10):
    container.upsert_item({
            'id': 'item{0}'.format(i),
            'TYkey' : '1',
            'productName': 'Widget',
            'productModel': 'Model {0}'.format(i)
        }
    )
```

```python
## Delete data
for item in container.query_items(query='SELECT * FROM emp e WHERE e.productModel = "Model 2"', enable_cross_partition_query=True):
    print(item)
    container.delete_item(item, partition_key=item['id'])
```

```python
## Query the database
for item in container.query_items(query='SELECT * FROM mycontainer r WHERE r.id in ("item3","item4")',enable_cross_partition_query=True):
    print(json.dumps(item, indent=True))
```

```python
import os
#from azure.cosmos import CosmosClient, exceptions, PartitionKey
from azure.cosmos.aio import CosmosClient 
import json
import asyncio

#URL = "https://rgrrcdb.documents.azure.com:443/"
#KEY = "u2sKgzsN40crDOZxwp1Nhaa1bCENhKdvetO45gp4ll0ZSSbu65grSaQQDQJhfJIhfotc0dEkdECOACDbAFxPUA=="
client = CosmosClient(URL, credential=KEY)
```

```python
## Using the asynchronous client
DATABASE_NAME='scott'
CONTAINER_NAME='emp'
async def create_products():
    async with CosmosClient(URL, credential=KEY) as client: # the with statement will automatically initialize and close the async client
        database = client.get_database_client(DATABASE_NAME)
        container = database.get_container_client(CONTAINER_NAME)
        for i in range(20):
            await container.upsert_item({
                    'id': 'item{0}'.format(i),
                    'productName': 'Widget',
                    'productModel': 'Model {0}'.format(i)
                }
            )
async def main():
    await create_products()
```

```python
# Run the event loop to execute the asynchronous function
await main()
#asyncio.run(main())
```
